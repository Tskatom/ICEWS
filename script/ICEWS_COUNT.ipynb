## Extract ICEWS count feature

```python
def extract_icews_feature(location, icews_folder=icews_count_folder):
    """
    return the ICEWS feature dataframe
    """
    series_list = []
    for fea_name in os.listdir(icews_folder):
        if fea_name == "20":
            continue
        fea_file = os.path.join(icews_folder, "%s/%s" % (fea_name, location.replace(" ", "_")))
        fea_series = pds.Series.from_csv(fea_file, sep='\t', index_col=0)
        fea_series.name = fea_name
        series_list.append(fea_series)
    p_date = pds.date_range('2014-04-01', '2014-12-07', freq='D')
    df = pds.concat(series_list, axis=1).sort_index(axis=1).reindex(p_date).fillna(0)
    return df
```

## Extract ICEWS intensity feature

```python
import json

icews_intency_folder = "/home/tskatom/workspace/icews_model/data/features/ICEWS/intencity/217/"
def extract_icews_intensity(location, icews_int_folder=icews_intency_folder):
    feature = {}
    for event_folder in os.listdir(icews_int_folder):
        if "20" == event_folder:
            continue
        loc_file = os.path.join(icews_int_folder, "%s/%s" % (event_folder, location.replace(" ","_")))
        with open(loc_file) as lf:
            for line in lf:
                day, obj_str = line.strip().split('\t')
                info = json.loads(obj_str)
                feature.setdefault(day, {"neg":{"count":0, "sum":0}, "pos":{"count":0, "sum":0}})
                feature[day]["neg"]["count"] += info["neg"]["count"]
                feature[day]["neg"]["sum"] += info["neg"]["sum"]
                feature[day]["pos"]["count"] += info["pos"]["count"]
                feature[day]["pos"]["sum"] += info["pos"]["sum"]
    #compute the daily everage
    average_fea = {"neg":{}, "pos":{}, "all":{}}
    for day in feature:
        neg_sum = 1.0*feature[day]["neg"]["sum"]
        neg_count = 1 if feature[day]["neg"]["count"] == 0 else feature[day]["neg"]["count"]
        pos_sum = 1.0*feature[day]["pos"]["sum"]
        pos_count = 1 if feature[day]["pos"]["count"] == 0 else feature[day]["pos"]["count"]
        
        average_fea["neg"][day] = round(neg_sum / neg_count, 3)
        average_fea["pos"][day] = round(pos_sum / pos_count, 3)
        average_fea["all"][day] = round((pos_sum + neg_sum) / (neg_count + pos_count), 3)
    
    #construct the dataframe
    df = pds.DataFrame(average_fea)
    p_date = pds.date_range('2014-04-01', '2014-12-07', freq='D')
    df.index = pds.DatetimeIndex(df.index)
    df = df.reindex(p_date).resample('W', how='mean')
    return df
```

## Extract Arabia Inform Feature

```python
arabia_feature_folder = "/home/tskatom/workspace/icews_model/data/features/ArabiaInform/related_doc/2014-12-12/"
def arabia_related_doc_fea(country, arabia_fea_folder=arabia_feature_folder):
    feature_list = ["protest", "coerce", "assault"]
    feature_dict = {}
    for feature in feature_list:
        feature_file = os.path.join(arabia_fea_folder, "%s.json" % feature)
        fh = open(feature_file)
        fea_json = json.load(fh).get(country, {})
        feature_dict[feature] = fea_json
    
    #construct the dataframe
    df = pds.DataFrame(feature_dict)
    p_date = pds.date_range('2014-04-01', '2014-12-07', freq='D')
    df.index = pds.DatetimeIndex(df.index)
    df = df.reindex(p_date).resample('W', how='sum').fillna(0)
    return df
```

## Combine ICEWS_COUNT, ICEWS_Intensity, Arabia Inform features | Construct Training Dataset

```python
merge_fea_folder = "/home/tskatom/workspace/icews_model/data/features/model/feature/"
target_folder = "/home/tskatom/workspace/icews_model/data/features/model/target/"
for country in mena_countries:
    icews_count_df = extract_icews_feature(country)
    icews_intensity_df = extract_icews_intensity(country)
    arabia_reldoc_df = arabia_related_doc_fea(country)
    final_feature = pds.concat([icews_count_df, icews_intensity_df, arabia_reldoc_df], axis=1).iloc[:-1]
    icews_target_count = icews_count_df[["14", "17", "18"]].iloc[1:]
    
    fea_out = os.path.join(merge_fea_folder, country.replace(" ", "_"))
    tar_out = os.path.join(target_folder, country.replace(" ", "_"))
    
    final_feature.to_csv(fea_out, sep='\t')
    icews_target_count.to_csv(tar_out, sep='\t')
    
```
