# Transform the weekly series into binary case(0 or 1)

```python
import matplotlib.pyplot as plt
import numpy as np
import json
import pandas as pd
import os

COUNTRIES = ["Egypt", "Iraq", "Bahrain", "Syria", "Libya", "Saudi Arabia", "Jordan"]
ICEWS_GSR_FOLDER = "/home/tskatom/workspace/icews_model/data/icews_gsr/206/"
EVENT_TYPE = ["14", "17", "18"]
start = "2012-01-01"
end = "2014-09-21"
num_countries = len(COUNTRIES)+1
for event_type in EVENT_TYPE:
    fig, axes = plt.subplots(nrows=num_countries/3 + 1, ncols=3,figsize=(20,10))
    for ix, country in enumerate(COUNTRIES):
        icews_file = os.path.join(ICEWS_GSR_FOLDER, "%s/%s" % (event_type, country.replace(" ","_")))
        icews_dict = {l.strip().split()[0]:int(l.strip().split()[1]) for l in open(icews_file)}
        icews_dict.setdefault(start, 0)
        icews_dict.setdefault(end, 0)
        
        icews_series = pd.Series(icews_dict)
        icews_series.index = pd.DatetimeIndex(icews_series.index)
        
        icews_weekly_series = icews_series.resample('W', how='sum').fillna(0)
        ax = axes[ix/3][ix%3]
        m = np.mean(icews_weekly_series)
        v = np.var(icews_weekly_series)
        icews_weekly_series[icews_weekly_series>0] = 1
        icews_weekly_series[icews_weekly_series==0] = -1
        ax.plot(icews_weekly_series,'bo',label="Pos(%d), Neg(%d)" % (sum(icews_weekly_series==1), sum(icews_weekly_series==-1)))
        ax.set_title(country)
        ax.legend()
    fig.suptitle('Event Type 06%s' % event_type)
    
```

# Construct a classifier to predict whether the weekly count will be zero or not

```python
import matplotlib.pyplot as plt
import numpy as np
import json
import pandas as pd
import os
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from datetime import timedelta
np.set_printoptions(threshold='nan')
COUNTRIES = ["Egypt", "Iraq", "Bahrain", "Syria", "Libya", "Saudi Arabia", "Jordan"]
#COUNTRIES = ["Bahrain"]
ICEWS_GSR_FOLDER = "/home/tskatom/workspace/icews_model/data/icews_gsr/206/"
HANDBOOK_DOCUMENT_FOLDER = "/home/tskatom/workspace/icews_model/data/handbook_document/"
HANDBOOK_KEYWORDD_FOLDER = "/home/tskatom/workspace/icews_model/data/handbook_words/"
EVENT_TYPE = ["14", "17", "18"]
#EVENT_TYPE = ["18"]
start = "2014-04-01"
end = "2014-09-14"
num_countries = len(COUNTRIES)+1
for event_type in EVENT_TYPE:
    for ix, country in enumerate(COUNTRIES):
        icews_file = os.path.join(ICEWS_GSR_FOLDER, "%s/%s" % (event_type, country.replace(" ","_")))
        icews_dict = {l.strip().split()[0]:int(l.strip().split()[1]) for l in open(icews_file) if l.strip().split()[0] >= start and l.strip().split()[0]<=end}
        icews_dict.setdefault(start, 0)
        icews_dict.setdefault(end, 0)

        #document count

        document_file = os.path.join('../data/handbook_document/', "%s_document_count.json" % country.replace(" ", "_"))
        document_obj = json.load(open(document_file))
        for day in document_obj.keys():
            if day > end or day < start:
                del document_obj[day]
        document_obj.setdefault(start, 0)
        document_obj.setdefault(end, 0)
        document_series = pd.Series(document_obj)
        document_series.index = pd.DatetimeIndex(document_series.index)
        document_series = document_series.resample('D').fillna(0)
        
        #keywords count
        keyword_file = os.path.join(HANDBOOK_KEYWORDD_FOLDER, "%s_keywords_count.json" % country.replace(" ", "_"))
        keywords_dict = json.load(open(keyword_file))
        for d in keywords_dict.keys():
            if d > end or d < start:
                del keywords_dict[d]
        days = sorted(keywords_dict.keys())
        data_list = [keywords_dict[d] for d in days]
        
        keywords_frame = pd.DataFrame(data_list, index=days)
        keywords_frame.index = pd.DatetimeIndex(keywords_frame.index)
        keywords_frame = keywords_frame.resample('D').fillna(0)
        
        
        
        
        start_test_index = -4
        
        icews_series = pd.Series(icews_dict)
        icews_series.index = pd.DatetimeIndex(icews_series.index)
        icews_series = icews_series.resample('D').fillna(0)
        
        icews_weekly_series = icews_series.resample('W', how='sum').fillna(0)
        binary_icews_weekly_series = icews_series.resample('W', how='sum').fillna(0)
        binary_icews_weekly_series[binary_icews_weekly_series > 0] = 1
        
        #using 8 weeks to as test
        trainY = binary_icews_weekly_series[2:start_test_index].values
        testY = binary_icews_weekly_series[start_test_index:].values
        
        #features
        """
        Previous two's icews weekly count
        previous two week's icews weekly binary count
        previous 5 day's daily count 
        """
        #trainX = np.c_[icews_weekly_series[:start_test_index-2],icews_weekly_series[1:start_test_index-1],binary_icews_weekly_series[:start_test_index-2],binary_icews_weekly_series[1:start_test_index-1]]
        trainX = np.c_[binary_icews_weekly_series[:start_test_index-2],binary_icews_weekly_series[1:start_test_index-1]]
        
        #get trainX for previos 5 day's icews daily count
        first_train_day = binary_icews_weekly_series.index[2]
        first_train_day_range = (first_train_day-timedelta(days=9),first_train_day-timedelta(days=7))
        num_train = len(trainY)
        start_t, end_t = first_train_day_range
        daily_features = []
        while num_train > 0:
            daily_values = list(icews_series[start_t:end_t].values)
            daily_features.append(daily_values)
            start_t, end_t = (start_t + timedelta(days=7), end_t + timedelta(days=7))
            num_train -= 1

        #Document feature
        first_train_day = binary_icews_weekly_series.index[2]
        first_train_day_range = (first_train_day-timedelta(days=6),first_train_day-timedelta(days=5))
        num_train = len(trainY)
        start_t, end_t = first_train_day_range
        doc_daily_features = []
        while num_train > 0:
            daily_values = list(document_series[start_t:end_t].values)
            doc_daily_features.append(daily_values)
            start_t, end_t = (start_t + timedelta(days=7), end_t + timedelta(days=7))
            num_train -= 1
        
        #keywords feature: we take first two days of keywords count as feature
        first_train_day = binary_icews_weekly_series.index[2]
        first_train_day_range = (first_train_day-timedelta(days=6),first_train_day-timedelta(days=5))
        num_train = len(trainY)
        start_t, end_t = first_train_day_range
        keywords_daily_features = []
        while num_train > 0:
            daily_values = list(np.sum(keywords_frame[start_t:end_t].values, axis=0))
            keywords_daily_features.append(daily_values)
            start_t, end_t = (start_t + timedelta(days=7), end_t + timedelta(days=7))
            num_train -= 1

        
        keywords_daily_features = np.array(keywords_daily_features)
        min_v = np.min(keywords_daily_features, axis=0)
        max_v = np.max(keywords_daily_features, axis=0)
        range_v = max_v - min_v
        keywords_daily_features = 1.0 * (keywords_daily_features - min_v) / range_v
        trainX = np.c_[trainX, keywords_daily_features]
        #print np.c_[trainX, trainY]
        
        #testX = np.c_[icews_weekly_series[start_test_index-2:-2], icews_weekly_series[start_test_index-1:-1], binary_icews_weekly_series[start_test_index-2:-2], binary_icews_weekly_series[start_test_index-1:-1]]
        testX = np.c_[binary_icews_weekly_series[start_test_index-2:-2], binary_icews_weekly_series[start_test_index-1:-1]]
        first_test_day = binary_icews_weekly_series.index[start_test_index]
        first_test_day_range = (first_test_day-timedelta(days=9),first_test_day-timedelta(days=7))
        num_test = len(testY)
        start_t, end_t = first_test_day_range
        daily_features = []
        while num_test > 0:
            daily_values = list(icews_series[start_t:end_t].values)
            daily_features.append(daily_values)
            start_t, end_t = (start_t + timedelta(days=7), end_t + timedelta(days=7))
            num_test -= 1
        
        #Document feature
        first_test_day = binary_icews_weekly_series.index[start_test_index]
        first_test_day_range = (first_test_day-timedelta(days=6),first_test_day-timedelta(days=5))
        num_test = len(testY)
        start_t, end_t = first_test_day_range
        doc_daily_features = []
        while num_test > 0:
            daily_values = list(document_series[start_t:end_t].values)
            doc_daily_features.append(daily_values)
            start_t, end_t = (start_t + timedelta(days=7), end_t + timedelta(days=7))
            num_test -= 1
            
        #keywords feature: we take first two days of keywords count as feature
        first_test_day = binary_icews_weekly_series.index[start_test_index]
        first_test_day_range = (first_test_day-timedelta(days=6),first_test_day-timedelta(days=5))
        num_test = len(testY)
        start_t, end_t = first_test_day_range
        keywords_daily_features = []
        while num_test > 0:
            daily_values = list(np.sum(keywords_frame[start_t:end_t].values, axis=0))
            keywords_daily_features.append(daily_values)
            start_t, end_t = (start_t + timedelta(days=7), end_t + timedelta(days=7))
            num_test -= 1
        
        keywords_daily_features = np.array(keywords_daily_features)
        keywords_daily_features = 1.0* (keywords_daily_features - min_v) /range_v
        
        testX = np.c_[testX, keywords_daily_features]
        
        #if all the training example is positive, then we predict all of them are positive
        if sum(trainY == 0) == 0:
            predictions = [1] * len(testY)
        else:
            """
            logReg = linear_model.LogisticRegression()
            logReg.fit(trainX, trainY)
            predictions = logReg.predict(testX)
            """
            #stupid prediction
            cursor = start_test_index
            predictions = []
            while cursor < 0:
                p_previous = binary_icews_weekly_series[cursor - 2]
                previous = binary_icews_weekly_series[cursor - 1]
                if p_previous == 0 and previous == 0:
                    predictions.append(0)
                elif p_previous == 0 and previous == 1:
                    predictions.append(1)
                elif p_previous == 1 and previous == 0:
                    predictions.append(0)
                else:
                    predictions.append(1)
                cursor += 1

        
        print "%s____06%s ---> Accurcy[%0.2f]" % (country, event_type, accuracy_score(testY, predictions)), testY, predictions
        #evaluate the performance using precision
    print "\n"
    
```
