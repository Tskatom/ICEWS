# Evaluate the ICEWS warnings

# Load icews warnings

```python
import boto
from datetime import timedelta
from dateutil import parser

db_conn = boto.connect_sdb()
domain = db_conn.lookup("warnings")
month_ranges = [("2014-07-01", "2014-08-01"), ("2014-08-01", "2014-09-01"), ("2014-09-01", "2014-10-01"),("2014-10-01", "2014-11-01"),("2014-11-01", "2014-12-01"),("2014-12-01", "2015-01-01")]
event_types = ["0614", "0617", "0618"]

initiate_warnings = []

for event_type in event_types:
    for month in month_ranges:
        start, end = month
        sql = "select * from warnings where mitreId != '' and eventDate >= '%s' and eventDate < '%s' and eventType='%s' " % (start, end, event_type)
        rs = domain.select(sql)
        initiate_warnings += [r for r in rs]

#clearne the warninigs being updated
#print len(initiate_warnings)
initiate_warnings = sorted(initiate_warnings, key=lambda x:x["date"], reverse=True)

updated_eid = []
cleaned_warnings = []
for warn in initiate_warnings:
    embers_update = warn["warningUpdate"]
    if embers_update == "None":
        if warn["embersId"] in updated_eid:
            continue
        else:
            cleaned_warnings.append(warn)
    else:
        updated_eid.append(embers_update)
        if warn["embersId"] in updated_eid:
            continue
        else:
            cleaned_warnings.append(warn)
#print len(cleaned_warnings)  

#construct the prediction dictionary
predictions = {}
for warn in cleaned_warnings:
    location = json.loads(warn["location"])
    country = location[0].encode('utf-8')
    state = location[1].encode('utf-8')
    city = location[2].encode('utf-8')
    if city == "-":
        if country == "Brazil":
            if state == "-":
                location_name = country
            else:
                location_name = country + "_" + state
        else:
            location_name = country
    else:
        location_name = country + "_" + city
    
    eventDate = warn["eventDate"]
    pred = int(warn["population"])
    event_type = warn["eventType"]
    weekend = (parser.parse(eventDate) + timedelta(days=4)).strftime("%Y-%m-%d")
    event_month = weekend[:7]
    
    predictions.setdefault(event_month, {})
    predictions[event_month].setdefault(location_name, {})
    predictions[event_month][location_name].setdefault(event_type, {})
    predictions[event_month][location_name][event_type][weekend] = pred
```

## Evaluate the Scores

```python
import numpy as np

#construct a location match
location_map = {"Brazil": "Brazil", "Venezuela_Caracas": "Caracas", "Libya_Tripoli": "Tripoli", "Bahrain_Manama":"Manama", "Jordan_Amman":"Amman", "Argentina":"Argentina",
 "Venezuela":"Venezuela", "Ecuador":"Ecuador", "Bahrain":"Bahrain", "Chile_Santiago":"Santiago", "Ecuador_Quito":"Quito", "Mexico":"Mexico", "El Salvador":"El_Salvador",
 "Argentina_Buenos Aires": "Buenos_Aires", "Iraq_Baghdad":"Baghdad", "Jordan":"Jordan", "Uruguay_Montevideo":"Montevideo", "Chile":"Chile", "Paraguay_Asunción":"Asuncion",
 "Iraq":"Iraq", "Mexico_Ciudad de México":"Mexico_City", "Libya":"Libya", "Uruguay":"Uruguay", "Syria":"Syria", "Syria_Damascus":"Damascus", "Egypt":"Egypt", "Egypt_Cairo":"Cairo",
 "El Salvador_San Salvador": "San_Salvador", "Colombia_Bogotá": "Bogota", "Saudi Arabia_Riyadh":"Riyadh", "Saudi Arabia":"Saudi_Arabia", "Colombia":"Colombia", "Paraguay":"Paraguay",
 "Brazil_Distrito Federal":"Distrito_Federal", "Brazil_Brasília": "Distrito_Federal"
}

def evaluate(actual, prediction):
    actual = max(0, actual)
    prediction = max(0, prediction)
    
    occurency = 0.5 * ((actual > 0) == (prediction > 0))
    accuracy = 3.5 * (1.0 - 1.0 * abs(actual - prediction) / max(actual, prediction, 4))
    return occurency + accuracy

def old_evaluate(actual, prediction):
    actual = max(0, actual)
    prediction = max(0, prediction)
    
    occurency = 2 * ((actual > 0) == (prediction > 0))
    accuracy = 2 * (1.0 - 1.0 * abs(actual - prediction) / max(actual, prediction, 1))
    return occurency + accuracy

#evaluate month by month
back_scores = {}
for month in predictions:
    for location in predictions[month]:
        icews_loc = location_map[location]
        for event_type in event_types:
            icews_et = event_type[2:]
            scores = []
            old_scores = []
            for weekend, count in predictions[month][location][event_type].items():
                score = evaluate(icews_gsr[icews_loc][icews_et].get(weekend, 0), count)
                scores.append(score)
                old_score = old_evaluate(icews_gsr[icews_loc][icews_et].get(weekend, 0), count)
                old_scores.append(old_score)
                
            mean_score = np.mean(scores)
            
            old_mean_score = np.mean(old_scores)
            back_scores.setdefault(month, {})
            back_scores[month].setdefault(location, {})
            back_scores[month][location][event_type] = {"new_score":mean_score, "old_score": old_mean_score}

#merge the score by country
merged_score_by_country = {}
for month in back_scores:
    for location in back_scores[month]:
        country = location.split("_")[0]
        for event_type in event_types:
            mean_score = back_scores[month][location][event_type]["new_score"]
            old_mean_score = back_scores[month][location][event_type]["old_score"]
            merged_score_by_country.setdefault(month, {})
            merged_score_by_country[month].setdefault(country, {})
            merged_score_by_country[month][country].setdefault(event_type, {"new_score":[], "old_score":[]})
            merged_score_by_country[month][country][event_type]["new_score"].append(mean_score)
            merged_score_by_country[month][country][event_type]["old_score"].append(old_mean_score)
            
```
