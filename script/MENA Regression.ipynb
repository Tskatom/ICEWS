```python
import os
import pandas as pds

MENA_COUNTRIES = ['Egypt', 'Jordan', 'Iraq', 'Libya', 'Bahrain', 'Syria', 'Saudi Arabia']
icews_count_folder = "/home/tskatom/workspace/icews_model/data/icews_counts/227"
MENA_counts = {}
for country in MENA_COUNTRIES:
    cf = os.path.join(icews_count_folder, "%s_icews_parent_event_counts_2012_227.csv" % country.lower().replace(" ", "_"))
    df = pds.DataFrame.from_csv(cf, sep='\t', index_col=1, infer_datetime_format=True)
    df = df.sort_index()["2012-01-01":'2015-02-15']
    df = df.resample('W', how='sum').fillna(0)
    MENA_counts[country] = df
    
```

```python
def evaluate(actual, prediction):
    actual = max(0, actual)
    prediction = max(0, prediction)
    
    occurency = 0.5 * ((actual > 0) == (prediction > 0))
    accuracy = 3.5 * (1.0 - 1.0 * abs(actual - prediction) / max(actual, prediction, 4))
    return round(occurency + accuracy, 3)
```

# Try Quantile Regression Model

```python
from sklearn.ensemble import GradientBoostingRegressor
trainIndex = [u'01', u'02', u'03', u'04', u'05', u'06', u'07', u'08', u'09', u'10', u'11', u'12', u'13', u'15', u'16', u'17', u'18', u'19', u'20']
testIndex = [u'14']
alpha=0.8

clf = GradientBoostingRegressor(loss='quantile', alpha=alpha,
                                n_estimators=250, max_depth=3,
                                learning_rate=.1, min_samples_leaf=9,
                                min_samples_split=9)
for country in MENA_counts:
    df = MENA_counts[country]
    N = len(df)
    T = 15
    X = df[trainIndex][0:N-T]
    y = df[testIndex][1:N-T+1]

    clf.set_params(alpha=1-alpha)
    clf.fit(X, y)
    xx = df[trainIndex][N-T:-9]
    yy = df[testIndex][N-T+1:N-8]
    # Make the prediction on the meshed x-axis
    y_pred = clf.predict(xx)
    fig = figure()
    plt.plot(range(1, T-8), yy.values, 'b.', label='real')
    plt.plot(range(1, T-8), floor(y_pred), 'r.', label='pred')
    plt.legend(loc='upper left')
    score = map(evaluate, yy.values, floor(y_pred))
    plt.title('%s_Score(%0.3f)' % (country, mean(score)))
    plt.show()
```
